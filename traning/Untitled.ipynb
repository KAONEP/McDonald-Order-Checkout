```python
from ultralytics import YOLO

if __name__ == "__main__":
    model = YOLO("yolo26n.pt")  

    model.train(
        data="data.yaml",
        epochs=200,        # 小数据适当多训
        imgsz=640,
        batch=16,          
        device=0,          # GPU
        workers=4,
        freeze=10,         # 冻结前10层（小数据很管用）
        lr0=0.002,         # 比默认更保守，减少发散
        patience=50,       # val 50轮不提升就早停
        cache=True,       # 小数据可直接缓存加速
        project="runs",
        name="yolo26_MCD"
    )
```

```python
model = YOLO("runs/yolo26_MCD/weights/best.pt")
metrics = model.val(data="data.yaml", split="test", imgsz=640, device=0)
print(metrics)
```
